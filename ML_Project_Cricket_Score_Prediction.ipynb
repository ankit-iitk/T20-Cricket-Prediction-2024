## ***Feature Engineering***


```python
# Replacing null values by 0 in [wides,noballs,byes,legbyes,penalty] columns

df['wides'].fillna(0,inplace = True)
df['noballs'].fillna(0,inplace = True)
df['byes'].fillna(0,inplace = True)
df['legbyes'].fillna(0,inplace = True)
df['penalty'].fillna(0,inplace = True)
```

```python
df['runs'] = df['runs_off_bat'] + df['extras'] + df['wides'] + df['noballs'] + df['byes'] + df['legbyes'] + df['penalty']
```

```python
# Drop unnecessary columns

df.drop(['striker','season','start_date','non_striker','runs_off_bat','extras','wides','noballs','byes','bowler','legbyes','penalty','wicket_type','other_wicket_type','other_player_dismissed'],axis = 1,inplace = True)
```

```python
# Take only 1st and 2nd innings dataset

df = df[(df['innings'] == 1) | (df['innings'] == 2)]
```

```python
# After removing 3rd and 4th innings shape of dataset

df.shape
```

```python
df['venue'].unique()
```

```python
# Extract city from venue column

df['city'] = df['venue'].apply(lambda x : x.split(', ')[-1])
```

```python
df
```

```python
df.drop('venue',axis = 1,inplace = True)
```

```python
# Create a new column based on runs column

df['current_score'] = df.groupby(['match_id','innings'])['runs'].cumsum()
```

```python
df
```

```python
# Extract over no. and ball no. from ball column

df['over'] = df['ball'].apply(lambda x:str(x).split('.')[0])
df['ball_no.'] = df['ball'].apply(lambda x:str(x).split('.')[1])
```

```python
df
```

```python
df['over'] = df['over'].astype(int)
df['ball_no.'] = df['ball_no.'].astype(int)
```

```python
# Creating balls_bowled column from over and ball_no.

df['balls_bowled'] = df['over']*6 + df['ball_no.']
```

```python
# Creating balls_left column from balls_bowled

df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
df
```

```python
df['player_dismissed'] = df['player_dismissed'].fillna('0')
df['player_dismissed'] = df['player_dismissed'].apply(lambda x : 0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype(int)
df['player_dismissed'] = df.groupby(['match_id','innings'])['player_dismissed'].cumsum()
```

```python
df['wicket_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
# Creating Current run rate

df['current_run_rate'] = (df['current_score']*6)/df['balls_bowled']
```

```python
groups = df.groupby(['match_id', 'innings'])
match_ids = df['match_id'].unique()

last_five = []

for id in match_ids:
  try:
    last_five.extend(groups.get_group((id, 1))['runs'].rolling(window=30).sum().values.tolist())
  except KeyError:
    pass

  try:
    last_five.extend(groups.get_group((id, 2))['runs'].rolling(window=30).sum().values.tolist())
  except KeyError:
    pass

df['last_five'] = last_five
```

```python
df
```

```python
# Total runs in each match

df = df.groupby(['match_id','innings']).sum()['runs'].reset_index().merge(df,on = ['match_id','innings'])
```

```python
df
```

```python
# Check null values

df.isnull().sum()
```

```python
# Drop null values

df.dropna(inplace = True)
```

# ***Model Building***

```python
# Separate x and y

x = final_df.drop('runs_x',axis = 1)
y = df['runs_x']
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)

print(f"x_train size : {x_train.shape}, x_test size : {x_test.shape}")
print(f"y_train size : {y_train.shape}, y_test size : {y_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# Define a function to find best fit model

from sklearn.model_selection import GridSearchCV

def best_model(x,y):
  models = {
      'Linear Regression': {
          'model': LinearRegression(),
          'parameters':{}
      },
      'Decision Tree Regressor':{
          'model': DecisionTreeRegressor(),
          'parameters':{'step2__criterion':['squared_error', 'friedman_mse'],'step2__splitter':['best','random'],'step2__max_depth':[5,10]}
      },
      'Random Forest Regressor':{
          'model': RandomForestRegressor(),
          'parameters':{'step2__criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],'step2__n_estimators':[1,2,10,50],'step2__max_depth':[5,10]}
      },
      'Support Vector Regressor':{
          'model': SVR(),
          'parameters':{'step2__C':[1,2,3,5],'step2__kernel':['linear','rbf','poly'],'step2__gamma':['scale','auto']}
      },
      'AdaBoost Regressor':{
          'model':AdaBoostRegressor(),
          'parameters':{'step2__n_estimators':[1,2,5,10],'step2__learning_rate':[0.01,0.1,1,5,10]}
      },
      'Gradient Boosting Regression':{
          'model': GradientBoostingRegressor(),
          'parameters':{'step2__n_estimators':[1,2,5,10],'step2__learning_rate':[0.01,0.1,1,5,10]}
      },
      'KNeighbors Regressor':{
          'model' : KNeighborsRegressor(),
          'parameters' :{'step2__n_neighbors':[1,3,5,7,9,11]}
      },
      'XGB Regressor':{
          'model' : XGBRegressor(),
          'parameters' : {'step2__n_estimators':[1,2,5,10],'step2__learning_rate':[0.01,0.1,1,5,10],'step2__max_depth':[5,10]}
      }

  }

  scores=[]
  for mod_name,mod_params in models.items():

    transformer = ColumnTransformer([
    ('OneHotEncoder',OneHotEncoder(),['batting_team','bowling_team','city']),
    ('StandardScaler',StandardScaler(),['innings','current_score','balls_left','wicket_left','current_run_rate','last_five'])
    ])

    pipe = Pipeline(steps=[
        ('step1',transformer),
        ('step2', mod_params['model'])
    ])

    print(f'Model - {mod_name}')
    print()
    print(mod_params['parameters'])


    classifier = GridSearchCV(estimator = pipe,param_grid = mod_params['parameters'],cv = 5,scoring='r2')
    classifier.fit(x,y)
    scores.append({
        'model':mod_name,
        'Best Parameters':classifier.best_params_,
        'Best Score':classifier.best_score_
    })
    print(f'Best Parameters - {classifier.best_params_}')
    print()
    print(f'R2 Score - {classifier.best_score_}')
    print()
    print('='*150)

  return pd.DataFrame(scores,columns = ['model','Best Parameters','Best Score'])
```

```python
pd.set_option('display.max_colwidth',170)
best_model(x_train,y_train)
```

Since KNeighbors Regressor and XGB Regressor are very close to 1 so to avoid overfitting, we will take Support Vector Regressor




```python
column_transformer = ColumnTransformer([
    ('OneHotEncoder',OneHotEncoder(),['batting_team','bowling_team','city']),
    ('StandardScaler',StandardScaler(),['innings','current_score','balls_left','wicket_left','current_run_rate','last_five'])
])
```

```python
pipeline = Pipeline(steps=[
    ('step1',column_transformer),
    ('step2',SVR(C=5,gamma='scale',kernel='rbf'))
])
```

```python
pipeline.fit(x_train,y_train)
```

## ***Model Evaluation***

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
y_pred = pipeline.predict(x_test)
```

```python
# Model Evaluation

print('Model Evaluation on Test Data\n')
print(f'R2 Score - {r2_score(y_test,y_pred)}')
print(f'Mean Absolute Error - {mean_absolute_error(y_test,y_pred)}')
print(f'Mean Squared Error - {mean_squared_error(y_test,y_pred)}')
print()

print('='*150)
print()

print('Model Evaluation on Train Data\n')
print(f'R2 Score - {r2_score(y_train,pipeline.predict(x_train))}')
print(f'Mean Absolute Error - {mean_absolute_error(y_train,pipeline.predict(x_train))}')
print(f'Mean Squared Error - {mean_squared_error(y_train,pipeline.predict(x_train))}')
```

```python
def predict_score(innings,batting_team,bowling_team,city,current_score,balls_left,wicket_left,current_run_rate,last_five):
  input_df = pd.DataFrame({'innings':[innings],'batting_team':[batting_team],'bowling_team':[bowling_team],'city':[city],'current_score':[current_score],'balls_left':[balls_left],'wicket_left':[wicket_left],'current_run_rate':[current_run_rate],'last_five':[last_five]})
  result = pipeline.predict(input_df)
  print(f'Predicted Score is {round(result.tolist()[0],3)}')
```
